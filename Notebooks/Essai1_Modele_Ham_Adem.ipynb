```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import csv
from datetime import datetime, timedelta

# Donnes spcifiques pour le Port San Luis, CA (hauteur d'eau moyenne, amplitudes, frquences, dphasages)
Z = 0.0  # Hauteur d'eau moyenne

# Fonction pour obtenir les donnes de mare spcifiques  une date donne
def get_tidal_data(date):
    csv_file = input("Enter the path to your CSV file: ")
    amplitudes = []
    frequencies = []
    phases = []

    with open(csv_file, 'r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            amplitude = float(row['Amplitude'])
            phase = float(row['Phase'])
            speed = float(row['Speed'])

            amplitudes.append(amplitude)
            frequencies.append(speed)
            phases.append(phase)

    return amplitudes, frequencies, phases



# Fonction de modle avec une somme de cosinus base sur les composantes de mare spcifiques
def tidal_model(t, Z, *parameters):
    result = Z
    for i in range(0, len(parameters), 3):
        Ac = parameters[i]
        Wc = 2 * np.pi * parameters[i + 1] / (3600 * 24)  # Conversion de la frquence en radians par jour
        Lc = parameters[i + 2]
        result += Ac * np.cos(Wc * t - Lc)
    return result



# test pour la prediction 
prediction_date = datetime(2024, 3, 11)
# Obtenir les donnes de mare pour la date spcifie
amplitudes, frequencies, phases = get_tidal_data(prediction_date)

# Gnrer les points temporels pour la prdiction
# Nous allons gnrer des points toutes les 10 minutes sur une journe
prediction_times = np.arange(0, 24 * 3600, 600)

# Initialiser les paramtres du modle pour l'ajustement
initial_parameters = [Z] + [c for component in zip(amplitudes, frequencies, phases) for c in component]

real_tide_data = np.zeros_like(prediction_times)

# Ajuster le modle aux donnes relles
fitted_parameters, _ = curve_fit(tidal_model, prediction_times, real_tide_data, p0=initial_parameters)

# Faire la prdiction
predicted_tide = tidal_model(prediction_times, *fitted_parameters)

# Afficher les rsultats
plt.figure(figsize=(10, 6))
plt.plot(prediction_times, predicted_tide, label='Prdiction de mare', linestyle='--', color='red')
plt.title('Prdiction de la hauteur d\'eau pour le ' + prediction_date.strftime('%d/%m/%Y'))
plt.xlabel('Temps (s)')
plt.ylabel('Hauteur d\'eau')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import csv
import os

class TidePredictionModel:
    def __init__(self):
        self.Z = 0.0  # Average water height
        self.temps = np.arange(0, 10, 0.01)  # Time from 0 to 10 seconds with a step of 0.01 second
        self.hauteur_eau = 3 * np.sin(2 * np.pi * 0.5 * self.temps) + 2 * np.sin(2 * np.pi * 2 * self.temps) + 0.5 * np.random.normal(size=len(self.temps))

    def extract_data_from_csv(self, csv_file):
        if not os.path.exists(csv_file):
            print(f"File {csv_file} does not exist.")
            return [], [], []

        amplitudes = []
        frequencies = []
        phases = []

        with open(csv_file, 'r') as file:
            csv_reader = csv.DictReader(file)
            for row in csv_reader:
                amplitude = float(row.get('Amplitude', 0))
                phase = float(row.get('Phase', 0))
                speed = float(row.get('Speed', 0))

                amplitudes.append(amplitude)
                frequencies.append(speed)
                phases.append(phase)

        return amplitudes, frequencies, phases

    def modele_mare(self, t, Z, *parametres):
        resultat = Z
        n = 0
        for i in range(0, len(parametres), 3):
            Ac = parametres[i]
            Wc = 2*3.14*(parametres[i + 1])/360
            Lc = parametres[i + 2]
            resultat += Ac * np.cos(Wc * (t/3600) - Lc)
            n += 1
        return resultat

    def predict_tide(self, csv_file):
        amplitudes, frequencies, phases = self.extract_data_from_csv(csv_file)

        print("Amplitudes:", amplitudes)
        print("Frequencies:", frequencies)
        print("Phases:", phases)

        
        subset_temps = self.temps[:120]
        subset_hauteur_eau = self.hauteur_eau[:120]
        
        
        #parametres_initiaux contenant les paramtres initiaux du modle de mare. Ces paramtres comprennent 
        #la hauteur d'eau moyenne self.Z et les paramtres d'amplitude, de frquence et de phase extraits du fichier CSV.
        parametres_initiaux = [self.Z] + [c for composante in zip(amplitudes, frequencies, phases) for c in composante]
        # Optimize parameters using curve_fit
        #Cette ligne utilise la fonction curve_fit de la bibliothque scipy.optimize pour ajuster les paramtres
        #du modle de mare aux donnes de hauteur d'eau observes. Les paramtres optimiss sont stocks dans parametres_optimaux.
        parametres_optimaux, _ = curve_fit(self.modele_mare, subset_temps, subset_hauteur_eau, p0=parametres_initiaux, maxfev=1000000)

        temps_prediction = np.arange(0, 24, 0.1)
        
        
        #Cette ligne utilise les paramtres optimiss pour prdire la hauteur d'eau
        # chaque instant de temps_prediction en appelant la mthode modele_mare avec les paramtres optimiss.
        hauteur_eau_predite = self.modele_mare(temps_prediction, *parametres_optimaux)

        # Plot results
        plt.figure(figsize=(10, 6))
        plt.plot(self.temps, self.hauteur_eau, label='Real Data')
        plt.plot(temps_prediction, hauteur_eau_predite, label='Tide Prediction', linestyle='--', color='red')
        plt.title('Prediction of water height with all tide constituents')
        plt.xlabel('Time (s)')
        plt.ylabel('Water Height')
        plt.legend()
        plt.show()


model = TidePredictionModel()
model.predict_tide("../Data/hc.csv")
```
